```python
# Check vocab of section annotation
data_fp1 = '/data1/longshen/Datasets/Piano/POP909/latents/song_level_phr_seq_with_annot_and_instconf/train_data.pt'
data_fp2 = '/data1/longshen/Datasets/Piano/POP909/latents/song_level_phr_seq_with_annot_and_instconf/valid_data.pt'
import torch
train_data = torch.load(data_fp1)
valid_data = torch.load(data_fp2)
data = {**train_data, **valid_data}
```

```python
data['001'].keys()
```

```python
n_sec = []
sec_type_set = set()
sec_len_set = set()
sec_annot_set = set()
for song_id in data:
    sec_annot = data[song_id]['phrase_annotation']
    n_sec.append(len(sec_annot))
    for (sec_type, sec_n_bar) in sec_annot:
        sec_annot_set.add(f"{sec_type}-{sec_n_bar}")
        sec_type_set.add(sec_type)
        sec_len_set.add(sec_n_bar)
```

```python
len(sec_annot_set)
```

```python
print(f'Section types: {sec_type_set}')
print(f'Section lengths: {sec_len_set}')
```

```python
print(f'Min number of sections: {min(n_sec)}')
print(f'Max number of sections: {max(n_sec)}')
```
