```python
# Find a good instrument condition in the dataset for generation in validation
from remi_z import MultiTrack
import torch
```

```python
tgt_n_bar_song_ids = []
for song_id in data:
    song = data[song_id]
    latent = song['latent']
    n_bar = latent.shape[0] // 4
    if n_bar == 64:
        tgt_n_bar_song_ids.append(song_id)
print(len(tgt_n_bar_song_ids))
print(tgt_n_bar_song_ids)
```
